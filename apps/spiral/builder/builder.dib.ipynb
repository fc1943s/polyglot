## types

```
inl types () =
    rust.types ()
    sm'.types ()
    file_system.types ()
    runtime.types ()
```

## get_args

```
inl get_args () =
    {
        file = "file"
        out_dir = "out-dir"
        args = "args"
    }
```

## get_command

```
let get_command () =
    ##"command"
    |> runtime.new_command
    |> runtime.command_arg (
        runtime.new_arg ##(get_args () .file)
        |> runtime.arg_short 'f'
        |> runtime.arg_long ##(get_args () .file)
        |> runtime.arg_required true
    )
    |> runtime.command_arg (
        runtime.new_arg ##(get_args () .out_dir)
        |> runtime.arg_short 'o'
        |> runtime.arg_long ##(get_args () .out_dir)
        |> runtime.arg_required true
    )
    |> runtime.command_arg (
        runtime.new_arg ##(get_args () .args)
        |> runtime.arg_short 'a'
        |> runtime.arg_long ##(get_args () .args)
    )
```

## run

```
let run { file out_dir args }
    : async.future_pin (resultm.result' bool sm'.std_string) =
    
    inl file = file |> file_system.get_full_path
    inl out_dir = out_dir |> file_system.get_full_path
    
    trace Debug
        fun () => $'"rust_builder.run"'
        fun () => $'$"file: {!file} / out_dir: {!out_dir} / args: {!args} / {!_locals ()}"'
    
    fun () =>
        inl result = true

        trace Debug
            fun () => $'"rust_builder.run"'
            fun () => $'$"result: {!result} / {!_locals ()}"'

        result |> Ok |> resultm.box
    |> async.future_init (2, 1) 1
```

```
// // test
// // rust=true
// // print_code=false

types ()
inl file = "file"
inl out_dir = "out_dir"
inl args = "args"

run {
    file
    out_dir
    args
}
|> async.block_on
|> resultm.unwrap'
|> sm'.format_debug'
|> sm'.from_std_string
|> _assert_eq (
    "true"
)
```

## tests

```
inl tests () =
    rust.run_tests [
        "verify_app", fun _ =>
            get_command () |> runtime.command_debug_assert
    ]
```

## main

```
// // rust=

inl main (args : array_base string) =
    trace Debug
        fun () => $'$"rust_builder.main"'
        fun () => $'$"args: {!args} / {!_locals ()}"'

    inl command = get_command ()
    inl arg_matches = command |> runtime.command_get_matches

    inl file =
        arg_matches
        |> runtime.matches_get_one (get_args () .file)
        |> optionm'.unbox
        |> optionm.value
        |> sm'.from_std_string

    inl out_dir =
        arg_matches
        |> runtime.matches_get_one (get_args () .out_dir)
        |> optionm'.unbox
        |> optionm.value
        |> sm'.from_std_string

    inl args =
        arg_matches
        |> runtime.matches_get_one (get_args () .args)
        |> optionm'.unbox
        |> optionm.map sm'.from_std_string

    run { file out_dir args }
    |> async.block_on
    |> resultm.unwrap'
    |> ignore

    0i32

inl main () =
    types ()
    $"let tests () = !tests ()" : ()
    $"let main args = !main args" : ()
```
